```python
clean_highlights = []
for i in dataset["train"]['summaries']:
    clean_highlights.append(i)
print("Summaries are complete.")

clean_article = []
for k in dataset["train"]['review_sents']:
    clean_article.append(k)
print("Texts are complete.")
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
model_name = "gpt2"  # You can choose different GPT-2 variants
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
```

```python
generated_summaries = []
for article in clean_article:
    input_text = f"Article: {article} Summarization:"

    # Encode the input text and set the pad_token_id to the end-of-sequence token
    input_ids = tokenizer.encode(input_text, return_tensors="pt", add_special_tokens=True, padding="max_length", max_length=150, truncation=True)
    attention_mask = input_ids != tokenizer.pad_token_id  # Create the attention mask

    # Generate the summary
    output = model.generate(input_ids, attention_mask=attention_mask, max_length=150, num_return_sequences=1, no_repeat_ngram_size=2)
    generated_summary = tokenizer.decode(output[0], skip_special_tokens=True)

    generated_summaries.append(generated_summary)

# Ensure the generated summaries are not empty
if not generated_summaries:
    print("No generated summaries found.")
else:
    # Print the generated summaries
    for summary in generated_summaries:
        print(summary)
```
